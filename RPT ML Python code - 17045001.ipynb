```python
#REGRESSION MODEL FOR Xp

f_xp = xp.drop(['xp'], axis = 1) #f_xp is feature dataset of SVR model
v_xp = xp['xp'] #v_xp is variable dataset
```

```python
#selecting only thousand files to ensure model don't over or underfit
f_xp = f_xp.iloc[0:1000].values 
v_xp = v_xp.iloc[0:1000].values
 
```

```python
#Scaling the data (0,1) as it is standard process in Ml models
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()

#Reshaing and transforming our data onto standard scale to prepare it for model
X = f_xp.reshape(-1,1)
X = sc_X.fit_transform(f_xp)
y = v_xp.reshape(-1,1)
y= sc_y.fit_transform(y)
```

```python
#Developing SVR Model 
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```
